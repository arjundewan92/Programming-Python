## Data Description

1. **com_id:**	Committee ID
2. **com_nam:**	Committee Name
3. **can_id:**	Candidate ID
4. **can_nam:**	Candidate name
5. **ele_yea:**	Election year
6. **can_off:**	Candidate Office
7. **can_off_sta:**	Candidate Office State
8. **can_off_dis:**	Candidate Office District
9. **lin_num:**	Line number from Detailed Summary Page of FEC Form 3
10. **lin_ima:**	Link to image presentation
11. **rec_com_id:**	Recipient Committee ID
12. **rec_nam:**	Recipient Name
13. **rec_str1:**	Recipient Street Address
14. **rec_str2:**	Apt. number or Suite number for recipient
15. **rec_cit:**	recipient city
16. **rec_sta:**	recipient state
17. **rec_zip:**	recipient zip
18. **dis_dat:**	date of disbursement
19. **dis_amo:**	amount of disbursement
20. **dis_pur_des:**	purpose description
21. **mem_cod:**	memo code
22. **mem_tex:**	memo text
23. **cat_cod:**	category code
24. **cat_des:**	category description
25. **tra_id:**	transaction id
26. **bac_ref_id:**	back reference id

#### Import required libraries and load the dataset

```python
import matplotlib.pyplot as plt

SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.style.use('ggplot')
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import seaborn as sns
```

# Data Visualization for 2016 Dataset

---

```python
df_2016 = pd.read_csv("D://Users/Shivam/Desktop/ArjunDallas/all_house_senate_2016.csv", encoding= "ISO-8859-1", low_memory=False)
```

```python
df_2016.dis_pur_des.value_counts().plot()
```

```python
df_2016.head(10)
```

```python
df_2016.shape
```

```python
df_2016.describe()
```

- Check missing values
- We will obtain the missing values as percentage of whole dataset.  

```python
df_2016.isnull().sum()/len(df_2016)*100.
```

#### Our data contains a lot of missing values

```python
counts = df_2016.groupby('can_off_sta').agg('count')['can_off']

plt.figure(figsize=(15,15))
plt.barh(range(len(counts)), counts, align = "center")
plt.yticks(range(len(counts)), counts.index)
plt.xlabel("Number of offices of Candidates in each State")
plt.title("Candidate Office State")
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.title("Candidate Office")
sns.countplot(x='can_off', data=df_2016)
```

```python
rec_sta1 = df_2016.rec_sta.value_counts()
rec_sta2 = rec_sta1.head(10)
plt.title("recipient state")
rec_sta2.plot(kind='bar',legend=True, figsize=(15,6), rot=15)
```

```python
dpd1 = df_2016.dis_pur_des.value_counts(ascending=False)
dpd2 = dpd1.head(10)
plt.title("Open-ended description of the purpose of the disbursement")
dpd2.plot(kind='pie', label="purpose description", legend=True, figsize=(15,6), rot=15)
```

```python
cat_cod1 = df_2016.cat_cod.value_counts(ascending=False)
cat_cod2 = cat_cod1.head(10)
plt.title("Number of counts of top 10 category code")
cat_cod2.plot(kind='line', label="category code", legend=True, figsize=(15,6), rot=15)
```

```python
cat_des1 = df_2016.cat_des.value_counts(ascending=False)
cat_des2 = cat_des1.head(10)
plt.title("Number of counts of top 10 category Description")
cat_des2.plot(kind='bar', label="category code", legend=True, figsize=(15,6), rot=70, colormap='Paired')
```
